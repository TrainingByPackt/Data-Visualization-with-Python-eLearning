## Flight passengers across years and months
In this activity, we will use a heatmap to find patterns in flight passenger data.

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mydata = pd.read_csv("flight_details.csv")
```

```python
mydata.head()
```

```python
data = mydata.pivot("Months", "Years", "Passengers")
```

```python
data
```

Use a heatmap to visualize the given data. The given dataset contains the monthly figures fpr flight passengers for multiple years. Use your own color map. Make sure that the lowest value is the darkest and the highest the brightest color.

```python
sns.set()
plt.figure(dpi=150)
```

```python
sns.heatmap(data, cmap=sns.light_palette("orange", as_cmap=True, reverse=True))
plt.title("Flight Passengers from 2001 to 2012")
#plt.xlabel("Years")
#plt.ylabel("Months")
plt.show()
```
